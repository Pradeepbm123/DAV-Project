```python
def plot_bar_graph(x_col, hue_col, title, xlabel, ylabel):
    plt.figure(figsize=(10, 5))
    sns.countplot(data=p, x=x_col, hue=hue_col)
    plt.title(title)
    plt.xticks(rotation=90)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(title=hue_col)
    plt.show()
```

```python
plot_bar_graph('Age', 'DietType', 'Distribution of Diet Types Across Age Groups', 'Age', 'Count')
```

```python
cross_table=pd.crosstab(p['Location'],p['PhysicalActivities'].count())
cross_table.plot(kind="bar")
```

```python
def plot_pie_chart(category_col, title):
    plt.figure(figsize=(7, 8))
    p[category_col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
    plt.title(title)
    plt.ylabel('')
    plt.show()
```

```python
plot_pie_chart('Gender', 'Gender Distribution')
```

```python
def plot_histogram(column, title, xlabel, bins=10):
    plt.figure(figsize=(10, 5))
    sns.histplot(p[column], bins=bins, kde=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plot_histogram('MonthlyElectricityConsumption', 'Distribution of Monthly Electricity Consumption', 'Monthly Electricity Consumption (kWh)')
```
